```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df_ori = pd.DataFrame({
    'Type': ['HMS','HMS','HMS','HMS','HMS','De-ID','De-ID','De-ID','De-ID','De-ID'],
    'Precision':[94.44,96.43,90.32,93.75,88.06,80,78.57,78.89,84.62,79.17],
    'Recall':[20.73,17.2,12.9,15,16.86,39.02,35.03,32.72,33,38],
    'Percentage':[90,80,70,60,50,90,80,70,60,50]
})

df_mod = pd.DataFrame({
    'Type': ['HMS','HMS','HMS','HMS','HMS','De-ID','De-ID','De-ID','De-ID','De-ID'],
    'Precision':[27.78,23.99,19.69,20.78,22.33,40.11,37.28,35.86,34.44,36.94],
    'Recall':[48.78,41.4,34.56,37.33,39.43,89.02,80.25,80.65,76.33,82.86],
    'Percentage':[90,80,70,60,50,90,80,70,60,50]
})

sns.set()
sns.set_style("whitegrid")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharex=True,figsize=(15,4))
c_map = sns.diverging_palette(10, 220, sep=80, n=5)
g1 = sns.barplot(y="Type", x="Precision", data=df_ori,palette=c_map,hue="Percentage",ax=ax1)
c_map = sns.color_palette("BrBG", 5)
g2 = sns.barplot(y="Type", x="Precision", data=df_mod,palette=c_map,hue="Percentage",ax=ax2)
g1.set_xlabel('# Precision on Original Scrubber')
g2.set_xlabel('# Precision on Modified Scrubber')
```
