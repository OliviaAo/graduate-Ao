```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Set palette
sns.set()
c_map = sns.diverging_palette(10, 220, sep=32, n=8)

# Set Dataset: Age:4, Date:1940, Doctor:879, Hospital: 626, ID:1083, Location:87, Patient:181, Phone:35
labels = ['Date', 'Doctor', 'Hospital', 'ID', 'Location', 'Patient', 'Phone','Age']
data = [1940, 879, 626, 1083, 87, 181, 35, 4]
```

```python
# Display Pie Chart
fig1, ax1 = plt.subplots()
explode = (0.0,0.0,0.0,0.0,0.3,0.5,0.7,0.9)
ax1.pie(data, labels=labels, autopct='%1.1f%%',colors=c_map,explode=explode,startangle=45)
ax1.axis('equal')
plt.show()
```
