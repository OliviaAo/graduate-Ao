```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df_ori = pd.DataFrame({
    'Type': ['HMS','HMS','HMS','HMS','HMS','De-ID','De-ID','De-ID','De-ID','De-ID'],
    'Precision':[0,0,0,0,0,27.27,5,8.51,5.83,7.48],
    'Recall':[0,0,0,0,0,4.41,1.8,3.24,2,2.56],
    'Percentage':[90,80,70,60,50,90,80,70,60,50]
})

df_mod = pd.DataFrame({
    'Type': ['HMS','HMS','HMS','HMS','HMS','De-ID','De-ID','De-ID','De-ID','De-ID'],
    'Precision':[9.25,10.64,10.05,8.4,9.9,3.78,4.12,3.16,3.05,4.48],
    'Recall':[55.88,44.91,34.82,28.86,27.67,22.06,18.56,10.53,10.57,16.74],
    'Percentage':[90,80,70,60,50,90,80,70,60,50]
})

sns.set()
sns.set_style("whitegrid")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True,figsize=(10,5))
c_map = sns.color_palette("RdBu_r", 5)
g1 = sns.barplot(x="Type", y="Precision", data=df_ori,palette=c_map,hue="Percentage",ax=ax1)
g2 = sns.barplot(x="Type", y="Precision", data=df_mod,palette=c_map,hue="Percentage",ax=ax2)
g1.set_xlabel('# Original Scrubber')
g2.set_xlabel('# Modified Scrubber')
```
