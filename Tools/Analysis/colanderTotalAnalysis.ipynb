```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df_ori = pd.DataFrame({
    'Type': ['HMS','De-ID','PHI-Reducer','MIST','Colander'],
    'Recall':[35.02,66.48,51.19,39.26,46.24],
})

sns.set()
sns.set_style("whitegrid")
c_map = sns.diverging_palette(10, 220, sep=32, n=5)
```

```python
fig, ax1 = plt.subplots()
ax1 = sns.barplot(x="Type", y="Recall", data=df_ori,palette=c_map)

plt.xlabel('# Different Scrubber')


ax2 = ax1.twinx()
x = [0,1,2,3,4]
p = df_ori['Recall']
ax2.plot(x,p,linestyle='--',c='black')

for a,b in zip(x,p):
    plt.text(a, b, '%.2f%%' %b, ha='center', va= 'bottom',fontsize=10)
ax2.set_yticks([]) 
plt.show()
```
