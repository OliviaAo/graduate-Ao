```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df_mist = pd.DataFrame({
    'Iteration':[5,5,5,5,5,10,10,10,10,10,15,15,15,15,15,20,20,20,20,20],
    'Precision':[55.38,27.24,28.76,65.26,54.74,54.47,46.45,62.46,55.26,67.74,61.63,51.09,58.74,58.99,65.46,59.55,59.72,54.01,57.05,63.17],
    'Recall':[41.86,40.22,39.02,36.68,45.28,38.95,43.25,38.82,40.31,46.2,30.81,38.84,30.98,35.7,37.56,30.81,34.71,30.39,35.01,33.99],
    'Percentage':[90,80,70,60,50,90,80,70,60,50,90,80,70,60,50,90,80,70,60,50] 
})
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.set()
sns.set_style("whitegrid")
c_map = sns.diverging_palette(10, 220, sep=32, n=4)
ax = sns.tsplot(time = 'Percentage',value='Precision',unit="Iteration",condition='Iteration',data=df_mist,color=c_map)
# ax = sns.pointplot(x="Percentage", y="Precision",hue="Iteration", data=df_mist, palette=c_map)
plt.show()
```
